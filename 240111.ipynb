```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
fig,axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('font',family='NanumGothic')
df = pd.read_excel('data1.xlsx')
plt.plot(df['name'],df['kor'],'k--',marker='o',label='국어점수',color='g')
plt.plot(df['name'],df['math'],marker='^',label='수학점수',color='r')
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('성적 그래프')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
iris=pd.read_csv("iris.csv")
print(iris)
```

```python
sns.scatterplot(x='petal_length',y='petal_width',hue='species',data=iris)
plt.title('Scatter Plot by seaborn',fontsize=20)
plt.show()
```

```python
sns.countplot(x='species',data=iris)
plt.show()
```

```python
sns.barplot(x='species',y='petal_length',data=iris,errorbar='sd')
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.show()
```

```python
sns.regplot(x='petal_length',y='petal_width',data=iris)
plt.show()
```

```python
df=iris.corr()
sns.heatmap(data=df,annot=True,fmt='.2f',cbar=False)
plt.show()
```
